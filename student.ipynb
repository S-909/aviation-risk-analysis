# 2. Data cleaning.
## 2.1 Filtering Data to Relevant Columns


Filtering Data to Relevant Columns

Modify `aviation_data` so that it only contains the relevant columns for our analysis. These columns are listed as a list of strings.

```python
# created a list with relevant columns for the analysis 
key_columns = ["Make","Model","Total.Uninjured","Aircraft.damage","Event.Date","Purpose.of.flight","Total.Fatal.Injuries","Total.Serious.Injuries","Total.Minor.Injuries"]
aviation_data = aviation_data[key_columns]
aviation_data
```

## 2.2 Filtering `Event_Data`.
Modifying `aviation_data` so that its column `Event_Data` data type changes from being an object to datetime object.

```python
#convert the Event.Date column from data type object to a datetime object.
aviation_data['Event.Date'] = pd.to_datetime(aviation_data['Event.Date'])
```

## 2.3 Changing data to lowercase.
The code below changes the object string in my data to be in lowercase and strips spaces.

```python
#making the column lower case and removing extra spaces.
txt_lower = aviation_data.select_dtypes(include='object').columns

aviation_data[txt_lower] = aviation_data[txt_lower].apply(lambda col: col.apply( lambda x: x.lower().strip() if isinstance(x, str) else x))
```

## 2.4 Create a New Column `Total people`

Creating a new column of the dataframe that represents the number of people who used airplanes and were involved in an accident. 

There is also use of `fillna()` to check `Total.fatal.injuries`,`Total.serious.injuries`,`Total minor injuries`,`Total.Uninjured` columns and fill the missing_values with 0.

The new column should be called `Total people` 

```python
# create a column fill the Nan values with 0 sum(axis=1) to Add across the columns for each 
aviation_data['Total_people'] = aviation_data[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']].fillna(0).sum(axis=1)
aviation_data['Total_people']
```

```python
#fill the missing values in Total.fatal.injuries with 0
aviation_data['Total.Fatal.Injuries'].fillna(0)
```

## 2.5 Drop Rows with Missing Values

Now that all rows in the dataset are key_columns, we need to drop all the rows with null values.

```python
#dropping rows with Nan values by using dropna()
aviation_data = aviation_data.dropna(subset=['Make', 'Model','Aircraft.damage', 'Purpose.of.flight'])
aviation_data.head()
```

# 3. Data Analysis.
Here there is grouping of the columns and calculating the fatality rate also we will be able to create visualizations using matplotlib and seaborn to be able to achieve our goals of getting;
- Which of the aircrafts make/model  has the lowest accident risks.
- Trend of accidents over time.
- Fatality rate versus purposes of flight.


## 3.1 Aircraft make/mode with the lowest accident risk.  
we will be grouping make and model column and create fatality rate column this will measure the overall fatality rate across all accidents
also we could look at the number of people dying per accident for each aircraft. 

```python
# grouped by make and model then agrregated the sum of Total.fatal.injuries and Total_people and counted the event date 
aircraft_risk = aviation_data.groupby(['Make', 'Model']).agg(
    Count_accident_occured =('Event.Date', 'count'),Total_fatal=('Total.Fatal.Injuries', 'sum'),Total_people_injured =('Total_people', 'sum')
).reset_index().copy()
#creating a new column called fatalities per accident that checks total fatalities per the number of times the accident occured
aircraft_risk['Fatalities_per_Accident'] = aircraft_risk['Total_fatal'] / aircraft_risk['Count_accident_occured']
#creating  a new column using aggreagated output tvo check total rate of fatalities
aircraft_risk['Fatalities_Rate'] = aircraft_risk['Total_fatal'] / aircraft_risk['Total_people_injured']
#drop duplicates
aircraft_risk_unique = aircraft_risk.drop_duplicates(subset= ['Fatalities_Rate','Fatalities_per_Accident'])
```

## 3.2 Trend of accident over time.  

This will help the management understand the trend of accidents over time

```python
#create a copy of the data
aviation_data = aviation_data.copy()
#create a new column called "year"
aviation_data['Year'] = aviation_data['Event.Date'].apply(lambda x: x.year)
#group by year and count accidents
accidents_per_year = aviation_data.groupby('Year').size().reset_index(name='Count_accident_occured')
```

## 3.3 Risk analysis by flight purpose.
This will help us understanding which of the two commercial flights or private flights has the lowest accident rates.

```python
#grouping the data using purpose of flight and give a summary statistics of count and sum of event date and total fatal injuries respectively.
flight_purpose = aviation_data.groupby('Purpose.of.flight').agg(
   Count_accident_occured =('Event.Date', 'count'),
     Total_fatal =('Total.Fatal.Injuries', 'sum')
).reset_index()
```

## 3.4 Aircraft damage analysis.


```python
aircraft_damage = aviation_data['Aircraft.damage'].value_counts().reset_index()
aircraft_damage.columns = ['Damage_Level', 'Count']
```

```python
aviation_data.to_csv("Cleaned_AviationData.csv", index=False)
```

# 4. Visualization.
Here i will be plotting bar graphs to show the top 10 low risk aircraft and the risk factor versus flight purpose.
Then a line graph to show accident trend  over time.


## 4.1 Importing `matplotlib` and `seaborn` libraries.


```python
#importing pyplot for plotting.
import matplotlib.pyplot as plt
#importing seaborn with its alias sns
import seaborn as sns
```

## 4.2 Bar chart showing the top 10 low risk aircraft 

```python
# Sort by fatalities per accident (ascending) to get low-risk aircraft
safe_aircraft = aircraft_risk.sort_values(by='Fatalities_per_Accident').head(10)

# Create a bar chart
plt.figure(figsize=(12, 6))
bars = plt.barh(
    safe_aircraft['Make'] + ' ' + safe_aircraft['Model'],
    safe_aircraft['Fatalities_per_Accident'],
    color='mediumseagreen'
)

plt.xlabel('Fatalities per Accident')
plt.title('Top 10 Low-Risk Aircraft (Lowest Fatalities per Accident)')
plt.gca().invert_yaxis()  # So the safest is at the top
plt.tight_layout()
plt.show()
```

## 4.3 Time series of accidents over the years.

```python
plt.figure(figsize=(12,6))
sns.lineplot(data = accidents_per_year, x='Year', y= 'Count_accident_occured')
plt.title('Aviation Accidents Over Time')
plt.xlabel('Year')
plt.ylabel('Count_accident_occured')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 4.4  Pie chart of Risk factor versus flight purpose.

```python
# Get top 5 most common purposes
top5_purposes = flight_purpose.sort_values('Count_accident_occured', ascending=False).head(5)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5_purposes['Count_accident_occured'], 
        labels=top5_purposes['Purpose.of.flight'], 
        autopct='%1.1f%%', 
        startangle=140)
plt.title('Top 5 Purposes of Flight by Accident Count')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## 4.5 Aircraft 

```python
plt.figure(figsize=(8,6))
sns.barplot(data= aircraft_damage, x='Damage_Level', y= flight_purpose['Total_fatal'], palette='magma')
plt.title('Aircraft Damage Levels')
plt.xlabel('Damage Level')
plt.ylabel('Total fatal')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
